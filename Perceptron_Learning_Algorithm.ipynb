```
import numpy as np 
import matplotlib.pyplot as plt
from scipy.spatial.distance import cdist
np.random.seed(2) #tạo random
means = [[2, 2], [4, 2]]
cov = [[.3, .2], [.2, .3]]
N = 10 # số lượng sample mỗi class
X0 = np.random.multivariate_normal(means[0], cov, N).T #sinh từng số trong class
X1 = np.random.multivariate_normal(means[1], cov, N).T

X = np.concatenate((X0, X1), axis = 1) #hợp nhất 2 class làm 1
y = np.concatenate((np.ones((1, N)), -1*np.ones((1, N))), axis = 1) #sinh output chuẩn
# Xbar 
X = np.concatenate((np.ones((1, 2*N)), X), axis = 0) #thêm biến số 1
def calculate_y(w, x):    
    return np.sign(np.dot(w.T, x)) #trả về ma trận dấu của w.T*x, nếu giá trị <0 thì trả về -1, >0 thì 1

def check(X, y, w):    
    return np.array_equal(calculate_y(w, X), y) #ktra xem 2 ma trận khi thực hiện thuật toán vs ma trận đầu ra có sẵn đã = nhau hay chưa
    
def PLA(X,y,w_init):
  w = [w_init]
  N = X.shape[1] #số lượng pt trong 1 class
  dimen = X.shape[0] #số lượng tham số (weight) trong hàm boundary
  miss = []
  iter = 0
  while True:
    iter += 1
    mix_data = np.random.permutation(N) #random 
    for i in range(N):
      xi = X[:,mix_data[i]].reshape(dimen,1) #lấy 1 tập hợp điểm (1 điểm mỗi class) để xét
      yi = y[0,mix_data[i]]
      if calculate_y(w[-1],xi)[0] != yi: #miss classiefied
        miss.append(mix_data[i]) #thêm giá trị ngoài vùng biên vào
        w_new = w[-1] + yi*xi #sử dụng SGD để tìm loss function tối ưu
        w.append(w_new) 
    if check(X,y,w[-1]): #check xem đã hội tụ hay chưa
      return (w[-1],len(miss))

dimen = X.shape[0]
w_init = np.random.randn(dimen, 1) #random phương trình 1 chiều
print(PLA(X,y,w_init))
```
